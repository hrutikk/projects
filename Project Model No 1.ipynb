```python
for column in x.columns:
    # Check if the column has missing values
    if x[column].isnull().any():
        if x[column].dtype == 'object':  # Check if column contains categorical data
            # Replace missing categorical values with mode
            mode_value = x[column].mode()[0]  # Calculate mode
            x[column].fillna(mode_value, inplace=True)
        else:
            # Replace missing numerical values with median
            median_value = x[column].median()  # Calculate median
            x[column].fillna(median_value, inplace=True)
```

# Pipeline

In a machine learning pipeline is the end-to-end construct that orchestrates the flow of data into, and output from, a machine learning model (or set of multiple models).

```python
cat=[]
con=[]
for column in x3.columns:
        if x3[column].dtype == 'object':  # Check if column contains categorical data
            cat.append(column)
        else:
            con.append(column)
```

```python
num_pipe=Pipeline(steps=[('imput',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encode',OneHotEncoder())])
```

```python
pre=ColumnTransformer([('cat_pipe',cat_pipe,cat),('num_pipe',num_pipe,con)])
pre
```

```python
x4=pd.DataFrame(pre.fit_transform(x3).toarray(),columns=pre.get_feature_names_out())
x4
```

# Split the Data

```python
x_train,x_test,y_train,y_test=train_test_split(x4,y,test_size=0.2,random_state=21)
```

# Evaluation of training and testing data

```python
lr.fit(x_train,y_train)
```

# Training Data

```python
y_pred_train=lr.predict(x_train)

mse=mean_squared_error(y_pred_train,y_train)
rmse=mse**0.5
mae=mean_absolute_error(y_pred_train,y_train)
r=r2_score(y_pred_train,y_train)

print('*'*20)
print('Mse:',mse)
print('RMSE:',rmse)
print('MAE:',mae)
print('R2 score:',r)
print('*'*20)
```

# Testing Data¶

```python
y_pred=lr.predict(x_test)

mse1=mean_squared_error(y_pred,y_test)
rmse1=mse**0.5
mae1=mean_absolute_error(y_pred,y_test)
r1=r2_score(y_pred,y_test)

print('*'*20)
print('Mse:',mse1)
print('RMSE:',rmse1)
print('MAE:',mae1)
print('R2 score:',r1)
print('*'*20)
```
